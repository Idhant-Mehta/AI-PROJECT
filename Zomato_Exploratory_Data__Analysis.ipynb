```python
final_df=pd.merge(df,df_country,on='Country Code', how='left')
```

```python
# to check datatypes
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_value=final_df.Country.value_counts().values
```

```python
# PIE CHART
plt.pie(country_value,labels=country_names)
```

### looking messed up so show top 3 countries

```python
plt.pie(country_value[:3],labels=country_names[:3],autopct='%1.2f%% ')
```

Observation: Zomato maximum records or transaction are from India. After that USA and then United Kingdom.

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
#reset_index() will convert it into df
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
#reset_index() will convert it into df
```

## Observations:

1. When rating is between 4.5 to 4.9 --> Excellent
2. When rating is between 4.0 to 4.4 --> Very Good
3. When rating is between 3.5 to 3.9 --> Good
4. When rating is between 2.5 to 3.4 --> Average
5. When rating is between 1.8 to 2.4 --> Poor
6. When rating 0.0 --> Not Rated

```python
ratings.head()
```

```python
import matplotlib as m
m.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
# use either " " or ' ' . no difference. both will work equally.
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings)
# not correct sequence of colors. so use palette command.
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
# white not visible so you can use blue instead or change theme.
```

## Observations:

1. Not rated count is very high.
2. Maximum number of rating are between 2.5 to 3.4

```python
# count plot .plotting w.r.t categorical Variable
#this tells us about how frequently it is coming. e.g. white has only one record. and red has 7 records etc.
sns.countplot(x='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
```
